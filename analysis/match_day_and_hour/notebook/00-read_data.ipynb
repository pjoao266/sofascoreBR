```python
import pandas as pd
```

```python
df_matches = pd.read_excel("../../../data/matches.xlsx")
df_teams = pd.read_excel("../../../data/team.xlsx")
```

```python
display(df_matches.head(3))
display(df_teams.head(3))
```

```python
df_matches = df_matches[["id_team_home", "id_team_away", "rodada", "date", "hour", "day_of_week", "city", "stadium"]]
df_teams = df_teams[["id", "name"]]
```

```python
df = pd.merge(df_matches, df_teams, how="left",  left_on='id_team_home', right_on='id').drop(columns="id").rename(columns={"name":"name_team_home"})
df = df[["id_team_home", "name_team_home", "id_team_away", "rodada", "date", "hour", "day_of_week", "city", "stadium"]]

df = pd.merge(df, df_teams, how="left",  left_on='id_team_away', right_on='id', suffixes=("", "_team_away")).drop(columns="id").rename(columns={"name":"name_team_away"})
df = df[["id_team_home", "name_team_home", "name_team_away", "rodada", "date", "hour", "day_of_week", "city", "stadium"]]
```

```python
df.to_csv("../data/processed/matches.csv", index=False)
```
